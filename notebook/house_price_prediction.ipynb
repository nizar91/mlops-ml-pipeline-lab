## 2 - Split data

```python
#  Split data into train, test using `train_test_split` with this paramaters: test_size=0.2, random_state=42, shuffle=True
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)
```

## 3 - train ml model

```python
# price houses
target_column = "MEDV"
```

```python
# get features for train
X_train = df_train.drop([target_column], axis=1)

X_train.head()
```

```python
# get target for train data
Y_train = df_train[target_column]

Y_train.head()
```

```python
# get feature for test data
X_test= df_test.drop(columns=[target_column], axis=1)

```

```python
# get target for test data
Y_test= df_test[target_column]

```

```python
# Create linear regression object
model = LinearRegression()

# Train the model using the training sets
## your code here
model.fit(X_train, Y_train)
```

```python
# Predict the values using the model on train data
train_predictions = model.predict(X_train)


# evaluate model on train data
print("Mean Squared Error on train data: \n")
mean_squared_error(Y_train, train_predictions)
```

```python
# Predict the values using the model on test data
test_predictions = model.predict(X_test)

# evaluate model on test data
print("Mean Squared Error on test data: \n")
## your code here))
print(mean_squared_error(Y_test, test_predictions))
```
